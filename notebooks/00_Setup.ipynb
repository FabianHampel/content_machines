# 🔗 ROTBOTS — Connect Notebooks via Google Drive

Run this cell **once** at the start of any notebook to set up the shared workspace.

All notebooks read and write to the same folder on your Google Drive,
so data flows automatically between them.

---

```python
# ============================================================
# SHARED SETUP — Run this in every notebook!
# Connects to Google Drive so all notebooks share the same files.
# ============================================================

from google.colab import drive
from pathlib import Path

# Mount Google Drive
drive.mount('/content/drive')

# Shared workspace on Google Drive
WORK_DIR = Path('/content/drive/MyDrive/rotbots_workshop')
WORK_DIR.mkdir(parents=True, exist_ok=True)
VIDEOS_DIR = WORK_DIR / 'videos'
VIDEOS_DIR.mkdir(exist_ok=True)

print(f'✅ Google Drive connected')
print(f'📁 Shared workspace: {WORK_DIR}')
print(f'🎬 Videos folder: {VIDEOS_DIR}')
print()
print('Files in workspace:')
for f in sorted(WORK_DIR.glob('*')):
    if f.is_file():
        print(f'   {f.name} ({f.stat().st_size / 1024:.0f}KB)')
```

After running this cell, all notebooks will use:

- `WORK_DIR` → `Google Drive/rotbots_workshop/`
- `VIDEOS_DIR` → `Google Drive/rotbots_workshop/videos/`

The Script Writer saves `prompts.json` there.  
The Video Generator reads it from there.  
Everything stays connected.
